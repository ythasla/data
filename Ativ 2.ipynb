Importando o tensorflow, o sklearn.cluster e as bibliotecas que serão usadas

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import AffinityPropagation
from tabulate import tabulate as tb
```

Importando os dados - numpy.load

```python
Goius = np.load('Goius.npy',allow_pickle=True)
Yehs = np.array(np.load('Yehs.npy',allow_pickle=True))
```

Visualização dos dados da primeira amostragem e análise de metricas estatísticas 

```python
plt.figure(figsize=(15,5))
plt.grid(True)
plt.plot(Goius[0])
plt.plot(Goius[1])
plt.show()
```

```python
print(tb([['Media: ', Goius[0].mean(), Goius[1].mean()], ['Variancia: ', Goius[0].var(), Goius[1].var()], 
          ['Amplitude: ', (Goius[0].max() - Goius[0].min()), (Goius[1].max() - Goius[1].min())], 
          ['Desvio Padrao: ', Goius[0].std(), Goius[1].std()]], headers=['Métrica', 'Goius 0', 'Goius 1']))
```

Visualização dos dados da segunda amostragem e análise de metricas estatísticas 

```python
plt.figure(figsize=(15,5))
plt.grid(True)
plt.plot(Yehs[0])
plt.plot(Yehs[1])
plt.show()
```

```python
print(tb([['Media: ', Yehs[0].mean(), Yehs[1].mean()], ['Variancia: ', Yehs[0].var(), Yehs[1].var()], 
          ['Amplitude: ', (Yehs[0].max() - Yehs[0].min()), (Yehs[1].max() - Yehs[1].min())], 
          ['Desvio Padrao: ', Yehs[0].std(), Yehs[1].std()]], headers=['Métrica', 'Yehs 0', 'Yehs 1']))
```

Using KMeans and Affinity Propagation

```python
X = Goius[0]
X = X.reshape(-1,1)
Kmean = KMeans(n_clusters=8, init = 'random')
X_pred = Kmean.fit(X)
print(X_pred)

clusteringX = AffinityPropagation().fit(X)
print(clusteringX)
```

```python
print(Kmean.cluster_centers_)
print(clusteringX.labels_)
```

```python
Kmean.labels_
```

```python
Y = Yehs[0]
Y = Y.reshape(-1,1)
Kmean = KMeans(n_clusters=8, init = 'random')
Y_pred = Kmean.fit(Y)
print(Y_pred)

clusteringY = AffinityPropagation().fit(X)
print(clusteringY)
```

```python
print(Kmean.cluster_centers_)
print(clusteringY.labels_)
```
